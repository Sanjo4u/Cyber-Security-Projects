```python
import matplotlib.pyplot as plt

# Data for the pie chart with respect to scenario preference
labels = ['Real World', 'Imaginary']

sizes = [15, 35] #15 chose real world, while 35 chose Imaginary scenario
colors = ['lightblue', 'lightgreen']
 

# Plotting the pie chart
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Engagement in Real-world vs. Imaginary Scenarios')
plt.axis('equal')  # Ensures that the pie chart is drawn as a circle
plt.show()
```
